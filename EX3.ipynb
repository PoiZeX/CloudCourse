```python
import ipywidgets as widgets
from ipywidgets import Layout, VBox
import IPython
from IPython.core.display import display_html
from typing import Text
from ipywidgets import widgets
from IPython.display import HTML
from google.colab.output import _publish as publish
from IPython.display import display
import pprint

# global vars / const
lst = []
students_dict = {}
currentStudentIndex = 0
filePath = 'students.json';

def readJsonToDict():
  with open(filePath, 'r') as file:
    global students_dict
    json_text = file.read()
    students_dict = json.loads(json_text)

def writeDictToJson():
  with open(filePath, 'w') as file: 
    file.write(strToWriteInJson) 
  file.close()

###################
## Define Fields ##
###################



style = {'description_width': '120px'} 

cb_students = widgets.Dropdown(
    options=lst,
    description='Student:',
    disabled=False,
    style=style,
)

txt_id = widgets.Text(
    description='ID:',
    disabled=True,
    style=style,

)

txt_mail = widgets.Text(
    description='Mail:',
    disabled=True,
    style=style,
)

txt_courses = widgets.Text(
    description='Courses:',
    disabled=True,
    style=style,
)

txt_link = widgets.Text(
    description='Link:',
    disabled=True,
    style=style,

)

txt_favShow = widgets.Text(
    description='Favorite show:',
    style=style,
   # layout = Layout(width='50%', height='80px', padding='10px')
)

button = widgets.Button(
    description='Update student',
    disabled=False,
    button_style='success', 
    layout=Layout(margin="20px 0px 0px 140px"),
    icon='check'
)

lb_success = widgets.Label(
    value="",
    layout=Layout(margin="20px 0px 0px 140px")
    )

def populateComboBox():
  for student in students_dict:
     lst.append(student["name"])
  cb_students.options = lst


##################
#### Behavior ####
##################

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        indx = 0
        for i in range(len(lst)):
          if lst[i] == change['new']:
            break
          indx += 1
        setupStudentToForm(indx) # initialize the selected student

def setupStudentToForm(studentIndex):
      global currentStudentIndex
      currentStudentIndex = studentIndex
      txt_id.value = students_dict[studentIndex]["id"]
      txt_mail.value = students_dict[studentIndex]["email"]
      txt_courses.value = students_dict[studentIndex]["courses"]
      txt_link.value = students_dict[studentIndex]["website"]
      if len(students_dict[studentIndex]) == 6:
        txt_favShow.value = students_dict[studentIndex]["fav show"]

def on_button_clicked(b):
    lb_success.value = ""
  # Display the message within the output widget.
    with open(filePath, 'w') as file: 
      students_dict[currentStudentIndex].update({"fav show" : txt_favShow.value});
      file.write(json.dumps(students_dict))
    lb_success.value = "Success"
    readJsonToDict() # refresh

###


####################
####### Main #######
####################

#setup the first student info
# writeDictToJson
def Main():
  writeDictToJson()  #-- uncomment to re-write the file if broken 
  
  # Read the data into dict
  readJsonToDict() # Read the
  
  populateComboBox()
  
  # set behaviors
  button.on_click(on_button_clicked)
  cb_students.observe(on_change)

  # setup the first student
  setupStudentToForm(currentStudentIndex) 
  
  # setup css stylesheet
  publish.css(" body {background-color: #77e1fc; display:flex; justify-content:center; "+
              "dropdown {background-color: #ffffff;  }")
  
  # put elemnts in screen
  output = widgets.Output()
  elements = [VBox([cb_students, txt_id, txt_mail, txt_courses, txt_link, txt_favShow, button, lb_success])]
  for element in elements:
    display(element, output)
  display(output)

Main() # start
```
